```python
import astropy.constants as ac #import a Python package to call the value of certain constants.
import astropy.units as u # import a Python package to call certain unit, e.g., the mass of the Sun
import numpy as np
```

how to define a function?

```python
def calculatingorbitalvelocity(M, R):
    """
    This is afunction that calculates the orbital velocity
    Input:
    M, the mass of the central object
    R, the orbital radius of the satellite
    Output:
    orbitalVelocity, the orbital velocity of the satellite
    """
    orbitalVelocity = np.sqrt(ac.G * M /R)
    return orbitalVelocity
```

```python
#Below we define input M and R
M= 1 * u.earthMass
R= 1 * u.earthRad
#Calculating orbital velocity of an earth satellite
v_earthSatellite = calculatingorbitalvelocity(M, R) # using the predefined function
print(v_earthSatellite.to(u.km/u.s))
```

We are going to introduce the concept of an array

```python
# Orbital radius of the eight solar system planets from inside out
# Mercury, venus, earth, mars, jupiter, saturn, uranus, neptune
R_planet_arr = np.array([0.4, 0.7, 1.0, 1.5, 5.2, 9.6,19.2, 30.1]) * u.au
```

```python
# Now we calculate the orbital velocities for eight planets
M_sun = 1 * u.solMass
v_planet_arr = calculatingorbitalvelocity(M_sun, R_planet_arr)
print(v_planet_arr.to(u.km/u.s))
```

```python
# prompt: can you help me with plotting v_planet_arr and R_planet_arr

import matplotlib.pyplot as plt # importing plotting package

plt.plot(R_planet_arr, v_planet_arr.to(u.km/u.s)) # Plotting velocity vs. orbital radius
plt.xlabel("Orbital Radius (AU)")# Add x axis label
plt.ylabel("Orbital Velocity (km/s)") # add y axis
plt.title("Orbital Velocity vs. Radius for Solar System Planets")
plt.show()
```

Below we will calculate the orbital velocity of a galaxy with only a central massive black hole - milky way -

```python
R_arr = np. arange(1, 30) * u.kpc # define radius of blackhole
print(R_arr.to(u.km))
```

```python
M_blackhole = 4.297 * 10**6 * u.Msun # define Mass of blackhole
print(M_blackhole)
```

```python
v_milky_arr = calculatingorbitalvelocity(M_blackhole, R_arr) # calcultae and define velocity of blackhole with def above
print(v_milky_arr.to(u.km/u.s))
```

```python

plt.plot(R_arr, v_milky_arr.to(u.km/u.s)) #
plt.xlabel('Distance from the center (kpc)') # label x axis
plt.ylabel('Orbital Velocity (km/s)') # label y axis
plt.title('Rotation Curve of a Galaxy with a Central Supermassive Black Hole') # label title
plt.show()
```
